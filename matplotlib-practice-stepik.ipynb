```python
def f(x, alpha, beta):
    return (x ** beta + alpha ** beta) / x ** beta
```

$f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta}}$

```python
import sys
import numpy as np
import matplotlib.pyplot as plt

print(sys.version)
print(np.__version__)
print(plt.matplotlib.__version__)
```

```python
alphas = [1, 2,]
betas = [1, 1]
```

### Задание 1

```python
plt.figure(figsize=(8, 6), dpi=100)

x = np.linspace(-5, 5, 100)
for alpha, beta in zip(alphas, betas):
    plt.plot(x, f(x, alpha, beta), label=rf'$ \alpha={alpha}, \beta={beta} $')

plt.title(r'График функции $f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta}}$ при различных $\alpha$ и $\beta$')
plt.xlim(-5, 5)
plt.ylim(-20, 30)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
```

Инсайты: ни какой график не находится в 4 четверти, при $\beta$ = 1 кривые находятся диаметрально противоположно, при $\beta$ = 2 с одной стороны относительно f(x), кажется что $\alpha$ влияет лишь на сами значения, но не на расположение линий, при $\alpha$ = 2, f(x) больше, что логично если смотреть на формулу.
