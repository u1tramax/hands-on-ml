```python
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
```

```python
a = 20
b = 45

x = np.linspace(a - 2, b + 2, 500)
probabilities = uniform.pdf(x, loc=a, scale=(b-a))
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, probabilities, color='skyblue')
plt.title(f'Unform distribution from {a} to {b}')
plt.xlabel('x')
plt.ylabel('Probability')
plt.tight_layout()
plt.grid(True)
plt.show()
```
