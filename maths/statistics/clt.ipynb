```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
means = []
samples = []
for i in range(10000):
    samples_poisson = np.random.exponential(scale=2, size=20)
    samples.append(samples_poisson)
    means.append(np.mean(samples_poisson))
```

```python
plt.subplot(1, 2, 1)
sns.histplot(np.array(samples).reshape(-1), kde=True, color='blue')
plt.title('Population Distribution (e.g., Exponential)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(means, kde=True, bins=100)
plt.title(f'Sampling Distribution of Means (n={10000})')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
