```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
n = 15
p = 0.2

k_values = np.arange(0, n+1)
probabilities = binom.pmf(k_values, n, p)
```

```python
plt.figure(figsize=(8, 6))
plt.bar(k_values, probabilities, color='skyblue', edgecolor='black')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title(f'Binomial distribution with n = {n} and p = {p}')
plt.xlabel('k')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.show()
```

```python
n = 1
p = 0.5

k_values = np.arange(0, n+1)
probabilities = binom.pmf(k_values, n, p)

plt.figure(figsize=(8, 6))
plt.bar(k_values, probabilities, color='skyblue', edgecolor='black')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title(f'Bernoulli distribution with n = {n} and p = {p}')
plt.xlabel('k')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.show()
```

25% of all students enrolled in high school XYZ are taking Algebra. 30 students are chosen at random. \
(a) Find the probability that exactly 7 students out of the 30 chosen are taking Algebra. \
(b) What is the probability that fewer than 5 students out of the 30 who are selected are taking Algebra? \
(c) Calculate the mean and standard deviation of this binomial distribution.

```python
from math import factorial as fact

p_algebra = 0.25
n = 30
result_a = (fact(30) / (fact(30 - 7) * fact(7))) * (p_algebra ** 7) * ((1 - p_algebra) ** (30 - 7))
result_a, binom.pmf(7, n, p_algebra)
```

```python
result_b = [
    (fact(30) / (fact(30 - x) * fact(x))) * (p_algebra ** x) * ((1 - p_algebra) ** (30 - x)) for x in range(1, 5)
]
result_b = np.sum(result_b)
result_b, binom.cdf(4, n, p_algebra)
```
