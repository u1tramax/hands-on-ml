```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

```python
mean = 28
k_values = np.arange(14, 42)

probabilities = poisson.pmf(k_values, mean)
```

```python
plt.figure(figsize=(8, 6))
plt.bar(k_values, probabilities, color='skyblue', edgecolor='black')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title(f'Poisson distribution with mean {mean}')
plt.xlabel('k')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.tight_layout()
plt.show()
```

A small business receives, on average, 8 calls per hour.\
(a) What is the probability that the business will receive exactly 7 calls in 1 hour? \
(b) What is the probability that the business will receive, at most, 5 calls in one hour? \
(c) What is the probability that the business will receive more than 6 calls in one hour?

```python
from math import factorial as fact

mean_calls = 8
result_a = mean_calls ** 7 * np.exp(-mean_calls) / fact(7)
result_a, poisson.pmf(7, mean_calls)
```

```python
result_b = [
    mean_calls ** x * np.exp(-mean_calls) / fact(x) for x in range(0, 6)
]
result_b = np.sum(result_b)
result_b, poisson.cdf(5, mean_calls)
```

```python
result_c = [
    mean_calls ** x * np.exp(-mean_calls) / fact(x) for x in range(0, 7)
]
result_c = 1 - np.sum(result_c)
result_c, poisson.sf(6, mean_calls)
```
