Найти вероятность того, что извлеченная наудачу деталь окажется отличного качества.

P(H2) - 2 завод, 20/50 \
P(H3) - 3 завод, 18/50 \

P(A|H1) = 0.9 \
P(A|H2) = 0.6 \
P(A|H3) = 0.9 \
P(A|H) - ? \
P(A|H) = P(A|H1)P(H1) + P(A|H2)P(H2) + P(A|H3)P(H3) = 12/50 * 9/10 + 20/50 * 6/10 + 18/50 * 9/10 = 108/500 + 120/500 + 162/500 = 390/500
