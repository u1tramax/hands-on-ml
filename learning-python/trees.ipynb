```python
from collections import deque
from typing import Optional, List

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def build_tree(values: List[Optional[int]]) -> Optional[TreeNode]:
    """
    values: список в level-order, None = отсутствующий узел
    Пример: [1, 2, 3, None, 4]
    """
    if not values or values[0] is None:
        return None

    root = TreeNode(values[0])
    queue = deque([root])
    i = 1

    while queue and i < len(values):
        node = queue.popleft()

        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1

        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1

    return root
```

```python
root = build_tree([1, 2, 3, 4, 5, None, 6])
```

```python
def leaves(root):
    leaves = []
    def dfs(root):
        if not root:
            return
        stack = [root]
        while stack:
            node = stack.pop()
            if not node.left and not node.right:
                leaves.append(node.val)
            # print(node.val)
            if node.right:
                stack.append(node.right)
            if node.left:
                stack.append(node.left)
    dfs(root)
    return leaves
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
import copy
from typing import Optional

class Solution:
    def pathSum(self, root: Optional[TreeNode], targetSum: int) -> int:
        cnt = 0
        setSums = set([0, root.val])
        def dfs(root, setSums):
            if root is None:
                return 0
            newSetSums = setSums.copy()
            cnt = 0
            if targetSum - root.val in setSums:
                cnt = 1
            newSetSums.add(root.val)
            print(newSetSums)
            return cnt + dfs(root.left, newSetSums) + dfs(root.right, newSetSums)
        return dfs(root, setSums)
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def longestZigZag(self, root: Optional[TreeNode]) -> int:
        self.result = 0
        def dfs(root, prev, sum):
            if root is None:
                return
            self.result = max(self.result, sum)
            if prev == 'left':
                dfs(root.right, 'right', sum+1)
                dfs(root.left, 'left', 1)
            else:
                dfs(root.left, 'left', sum+1)
                dfs(root.right, 'right', 1)
        dfs(root, 'left', 0)
        dfs(root, 'right', 0)
        return self.result
```
