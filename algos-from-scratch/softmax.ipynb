```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class SoftmaxRegression:
    
    learning_rate = 0.1
    iterations = 1000

    def fit(self, X, y):

        m, n = X.shape
        X = np.c_[np.ones(m), X]
        n += 1
        theta = np.zeros((n, len(np.unique(y))))

        def softmax(X):
            return np.exp(X.T @ theta) / np.sum(np.exp(X.T @ theta), axis=1)
        
        def entropy_gradient(X, y):
            return (1 / m) * ((softmax(X) - y) @ X)

        for _ in range(self.iterations):
            theta -= self.learning_rate * entropy_gradient(X, y).T

        self.theta = theta
        print(theta)

    def predict_proba(self, X):
        
        return np.exp(X @ self.theta) / np.sum(np.exp(X @ self.theta), axis=1)
    
    def predict(self, X):

        return np.argmax(self.predict_proba(X), axis=1)
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
X, y = iris['data'], iris['target']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```
