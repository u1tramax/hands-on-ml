```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class MyLinearRegression():


    def __init__(self, learning_rate: float, n_iterations: int):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations

    def fit(self, X: pd.DataFrame, y: pd.DataFrame):

        m, n = X.shape
        X = np.c_[np.ones(m), X]
        n += 1
        theta = np.zeros((n))

        def gradient_mse(X, y, theta):
            return 1 / m * X.T.dot(X.dot(theta) - y)
        
        for _ in range(self.n_iterations):
            theta -= self.learning_rate * gradient_mse(X, y, theta)

        self.theta = theta

    def predict(self, X):

        X = np.c_[np.ones(X.shape[0]), X]
        return X.dot(self.theta)
```

```python
learning_rate = 0.1
iterations = 10000
X = np.array([[5, 2], [7, 2], [2, 3]])
y = np.array([1, 3, 2])
mylinreg = MyLinearRegression(learning_rate, iterations)
mylinreg.fit(X, y)
```

```python
mylinreg.predict(np.array([[5, 2], [7, 2], [2, 3]]))
```
