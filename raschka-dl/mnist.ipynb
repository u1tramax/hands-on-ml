```python
from sklearn.datasets import fetch_openml
```

```python
X, y = fetch_openml('mnist_784', version=1, return_X_y=True)
```

```python
X = X.values
y = y.astype(int).values
```

```python
X, y
```

```python
X.shape, y.shape
```

```python
X = ((X / 255) - .5) * 2
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=2, ncols=5, sharex=True, sharey=True)
ax = ax.flatten()

for i in range(10):
    img = X[y == i][0].reshape(28, 28)
    ax[i].imshow(img, cmap='Greys')

ax[0].set_xticks([])
ax[0].set_yticks([])
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=5, ncols=5, sharex=True, sharey=True)
ax = ax.flatten()

for i in range(25):
    img = X[y == 7][i].reshape(28, 28)
    ax[i].imshow(img, cmap='Greys')

ax[0].set_xticks([])
ax[0].set_yticks([])
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=15000, random_state=123, stratify=y)
X_val, X_test, y_val, y_test = train_test_split(X_val, y_val, test_size=10000, random_state=123, stratify=y_val)
```
